### Passo 1 - Importar Arquivos e Bibliotecas

```python
import pandas as pd
import pathlib
import win32com.client as win32
```

```python
email = pd.read_excel(r'C:\Users\leona\Desktop\Python\Projetos\Projeto 1\Projeto Automacao Indicadores\Bases de Dados\Emails.xlsx')
lojas = pd.read_csv(r'C:\Users\leona\Desktop\Python\Projetos\Projeto 1\Projeto Automacao Indicadores\Bases de Dados\Lojas.csv', encoding='latin1', sep=';')
vendas = pd.read_excel(r'C:\Users\leona\Desktop\Python\Projetos\Projeto 1\Projeto Automacao Indicadores\Bases de Dados\Vendas.xlsx')
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
display(email)
```

```python
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
dic_vendas = {}
for loja in lojas['Loja']:
    dic_vendas[loja] = vendas.loc[vendas['Loja']==loja,:]
display(dic_vendas['Iguatemi Esplanada'])
```

```python
dia_indicador = vendas['Data'].max()
print(dia_indicador)
print(f'{dia_indicador.day}/{dia_indicador.month}')
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
#local das pastas
caminho_backup = pathlib.Path(r'C:\Users\leona\Desktop\Python\Projetos\Projeto 1\Projeto Automacao Indicadores\Backup Arquivos Lojas')
arquivos_pasta_backup = caminho_backup.iterdir()
lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]

#criar a pasta, caso não exista
for loja in dic_vendas:
    if loja not in lista_nomes_backup:
        Path(caminho_backup / loja).mkdir()
    #criar os arquivos dentro das pastas
    nome_arquivo = f'{dia_indicador.day}_{dia_indicador.month} {loja}.xlsx'
    local_arquivo = caminho_backup / loja / nome_arquivo
    dic_vendas[loja].to_excel(local_arquivo) 
```

### Passo 4 - Calcular o indicador para 1 loja

```python
#difinir metas

meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_diversidade_dia = 4
meta_diversidade_ano = 120
meta_ticket_dia = 500
meta_ticket_ano = 500
```

```python
for loja in dic_vendas:
    vendas_loja = dic_vendas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data']==dia_indicador, :]

    #Faturamento
    faturamento_anual = vendas_loja['Valor Final'].sum()
    faturamento_dia = vendas_loja_dia['Valor Final'].sum()

    #Diversidade
    diversidade_anual = len(vendas_loja['Produto'].unique())

    diversidade_dia = len(vendas_loja_dia['Produto'].unique())

    #Ticket medio
    valor_venda_anual = vendas_loja.groupby('Código Venda').sum()
    ticket_anual = valor_venda_anual['Valor Final'].mean()

    valor_venda_dia = vendas_loja_dia.groupby('Código Venda').sum()
    ticket_dia = valor_venda_dia['Valor Final'].mean()
    outlook = win32.Dispatch('outlook.application')

    #Enviar email para gerencia
    mail = outlook.CreateItem(0)
    mail.TO = email.loc[email['Loja']==loja, 'E-mail'].values[0]
    mail.Subject = f'OnePage Dia {dia_indicador.day}/{dia_indicador.month}/{dia_indicador.year} - {loja}'

    if faturamento_dia >= meta_faturamento_dia:
        cor_fat_dia = "green"
    else:
        cor_fat_dia = "red"
    if faturamento_anual >= meta_faturamento_ano:
        cor_fat_ano = "green"
    else:
        cor_fat_ano = "red"
    if diversidade_dia >= meta_diversidade_dia:
        cor_diver_dia = "green"
    else:
        cor_diver_dia = "red"
    if diversidade_anual >= meta_diversidade_ano:
        cor_diver_ano = "green"
    else:
        cor_diver_ano = "red"
    if ticket_dia >= meta_ticket_dia:
        cor_ticket_dia = "green"
    else:
        cor_ticket_dia = "red"
    if ticket_anual >= meta_ticket_ano:
        cor_ticket_ano = "green"
    else:
        cor_ticket_ano = "red"

    mail.HTMLBody = f'''
    <p>Bom dia, {email.loc[email['Loja']==loja, 'Gerente'].values[0]}</p>
    <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>loja {loja}</strong> foi:</p>

    <table border="1">
      <tr>
        <th>Indicador</th>
        <th>Valor dia</th>
        <th>Meta dia</th>
         <th>Cenário dia</th>
      </tr>
      <tr>
        <td style="text-align: center">Faturamento</td>
        <td style="text-align: center">R${faturamento_dia:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
      </tr>
      <tr>
        <td style="text-align: center">Diversidade de produtos</td>
        <td style="text-align: center">{diversidade_dia}</td>
        <td style="text-align: center">{meta_diversidade_dia}</td>
        <td style="text-align: center"><font color="{cor_diver_dia}">◙</font></td>
      </tr>
      <tr>
        <td style="text-align: center">Ticket médio</td>
        <td style="text-align: center">R${ticket_dia:.2f}</td>
        <td style="text-align: center">R${meta_ticket_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
      </tr>
    </table>
    <br>
    <table border="1">
      <tr>
        <th>Indicador</th>
        <th>Valor dia</th>
        <th>Meta dia</th>
         <th>Cenário dia</th>
      </tr>
      <tr>
        <td style="text-align: center">Faturamento</td>
        <td style="text-align: center">R${faturamento_anual:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
      </tr>
      <tr>
        <td style="text-align: center">Diversidade de produtos</td>
        <td style="text-align: center">{diversidade_anual}</td>
        <td style="text-align: center">{meta_diversidade_ano}</td>
        <td style="text-align: center"><font color="{cor_diver_ano}">◙</font></td>
      </tr>
      <tr>
        <td style="text-align: center">Ticket médio</td>
        <td style="text-align: center">R${ticket_anual:.2f}</td>
        <td style="text-align: center">R${meta_ticket_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
      </tr>
    </table>

    <p>Segue em anexo a planilha com todos os dados para mais detalhes.</p>
    <p>Qualquer duvida estou a disposição</p>
    <p>Atenciosamente, Leonardo</p>
        '''

    attachment = r'C:\Users\leona\Desktop\Python\Projetos\Projeto 1\Projeto Automacao Indicadores\Backup Arquivos Lojas\{}\{}'.format(loja, f'{dia_indicador.day}_{dia_indicador.month} {loja}.xlsx')
    mail.Attachments.Add(attachment)
    mail.Send()
```

### Passo 5 - Criar ranking para diretoria

```python
faturamento_lojas = vendas.groupby('Loja')[['Loja','Valor Final']].sum()
faturamento_lojas = faturamento_lojas.sort_values(by='Valor Final', ascending=False)

nome_arquivo = f'{dia_indicador.day}_{dia_indicador.month} Ranking Anual'
faturamento_lojas.to_excel(r'Backup Arquivos Lojas\{}.xlsx'.format(nome_arquivo))

vendas_dia = vendas.loc[vendas['Data']==dia_indicador,:]
faturamento_lojas_dia = vendas_dia.groupby('Loja')[['Loja','Valor Final']].sum()
faturamento_lojas_dia = faturamento_lojas_dia.sort_values(by='Valor Final', ascending=False)

nome_arquivo = f'{dia_indicador.day}_{dia_indicador.month} Ranking Diario'
faturamento_lojas_dia.to_excel(r'Backup Arquivos Lojas\{}.xlsx'.format(nome_arquivo))
```
